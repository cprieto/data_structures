We need a simple method to create a node, for this we can create a simple helper function.

```c
#include <stdio.h>
#include <stdlib.h>

typedef struct node {
    int num;
    struct node *next;
} Node, *NodePtr;

NodePtr makeNode(int);

int main() {
    NodePtr node = makeNode(1);
    node->next = makeNode(2);
    node->next->next = makeNode(3);
    
    printf("3rd node is %i\n", node->next->next->num);
}

NodePtr makeNode(int num) {
    NodePtr node = malloc(sizeof(Node));
    node->num = num;
    node->next = NULL;
    return node;
}
```

As we see it is simple, we move through the list until the next is `NULL`.
